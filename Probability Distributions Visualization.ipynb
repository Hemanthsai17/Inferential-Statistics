# Visualization of Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

## Binomial

```python
def binomial(n,p):
    x = np.arange(n)
    def prob(x):
        return(math.comb(n,x)*(p**x)*(1-p)**(n-x))
    y = []
    for i in range(len(x)) :
        y.append(prob(i))

    plt.bar(x,y)
    plt.title(" binomial distribution")
    plt.xlabel("X")
  
    
    plt.show()

binomial(50,0.5)
```

## Normal

```python
def normal(mu,sig):
    p=(mu-(10*sig*sig))
    q=(mu+(10*sig*sig))
    x = np.linspace(p,q,100)
    def prob(x):
        k=(1/((sig)*math.pow((2*math.pi),0.5)))
        l=(-1)/(2*sig*sig)
        m=math.pow((x-mu),2)
        j=(math.exp(l*m))
        return(j)
    y=[]
    for i in range(len(x)) :
        y.append(prob(i))
    
    plt.plot(x,y)
normal(50,5)
```

## Exponential

```python
def exponential(lamb):
    x = np.linspace(-10,10,100)
    def prob(x):
        j=(math.exp((-1)*lamb*x))
        return(lamb*j)
    y=[]
    for i in range(len(x)) :
        y.append(prob(i))
    
    plt.plot(x,y)
    plt.xlim(-10,5)
exponential(0.1)
```

## Poisson

```python
def poisson(lamb):
    x = np.linspace(-10,10,100)
    def prob(x):
        l=math.pow(lamb,x)
        m=math.factorial(x)
        n=math.exp((-1)*lamb)
        j=((l/m)*n)
        return(j)
    y=[]
    for i in range(len(x)) :
        y.append(prob(i))
    
    plt.bar(x,y)
    plt.plot(x,y,color='red')
    plt.xlim(-11,-5)
poisson(5)
```
